Load the LLM (LLaMA-3.2)

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Specify the Llama 3.2 model
model_name = "meta-llama/Llama-3.2-3B"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

print("Llama 3.2 Model Loaded Successfully!")
```

Load & Process Documents

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load and split PDF into smaller chunks
loader = PyPDFLoader("/content/Candidate-Handbook-February-12-2025-cover-com.pdf")
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
docs = text_splitter.split_documents(documents)
```

Generate Embeddings & Store in FAISS

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# Use Sentence Transformers for text embeddings
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Convert docs to embeddings & store in FAISS
vector_store = FAISS.from_documents(docs, embedding_model)
vector_store.save_local("faiss_index")
```

 Retrieve

```python
from langchain.chains import RetrievalQA
from langchain_huggingface import HuggingFacePipeline
from langchain.vectorstores import FAISS
from transformers import pipeline

import torch

# Ensure Model Uses GPU if Available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load FAISS index with correct embedding function
#vector_store = FAISS.load_local("faiss_index", embeddings=embedding_model, allow_dangerous_deserialization=True)

# Set up retriever with limited search results
retriever = vector_store.as_retriever(search_kwargs={"k": 10})  # Reduce retrieved documents



# Define LLM pipeline correctly (with explicit device setting & max tokens)
llm_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=100,
    truncation=True,
    do_sample=True,
    temperature=0.7,
    pad_token_id=tokenizer.eos_token_id  # Avoids unnecessary warnings
)

# Define RAG pipeline
rag_chain = RetrievalQA.from_chain_type(
    llm=HuggingFacePipeline(pipeline=llm_pipeline),
    retriever=retriever
)
```

Generate Response

```python
# Query
query = "What are Consumer Benefits"
response = rag_chain.invoke(query)

# Format and Print Output
#formatted_output = response['result'].replace("•", "-").replace("\n\n", "\n")  # Standardize bullet points and spacing
# Process Output
formatted_output = response['result'].replace("•", "-").replace("\n\n", "\n").strip()

print("\n=== Testing Environment Setup Requirements ===\n")
print(formatted_output)
```
